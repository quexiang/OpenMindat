This Notebook mainly shows how to retrieve locality records based on country, locality ID, and descriptions, including the following 5 examples:

1. Retrieve records of locality in a specified country, such as China. 
2. Retrieve records of locality by a specified id. 
3. Retrieve records of locality by giving a specified description. 


```r
library(httr)
library(jsonlite)
library (OpenMindat)
```

```r
#Please replace the following string “9ce67655d74bcd981e937be80dcea9cb” by your own Mindat token.
mindat_connection("9ce67655d74bcd981e937be80dcea9cb",page_size = 1500)
```

1. Retrieve records of locality in a specified country, such as China:

```r
df1<-localities_list_country ("China")
```

```r
df1
```

2. Retrieve records of locality by a specified id:

```r
df2<-localities_retrieve_id(id = 22)
```

```r
df2
```

3. Retrieve records of locality by giving a specified description, such as "volcano":

```r
df3<-localities_list_description("volcano")
```

```r
df3
```
