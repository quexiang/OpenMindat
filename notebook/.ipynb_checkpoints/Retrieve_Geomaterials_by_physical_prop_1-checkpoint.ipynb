This Notebook mainly shows how to retrieve geomaterials records based on physical properties with numerical values or threshold ranges, including the following 5 examples:

1. Retrieve records of geomaterials with a Mohs hardness greater than 9. 
2. Retrieve records of geomaterials with a Mohs hardness of less than 1. 
3. Retrieve records of geomaterials with a Mohs hardness ranging from 3 to 3.5.
4. Retrieve records of geomaterials with a density ranging from 3 to 3.2.
5. Retrieve records of geomaterials with a calculated 2V angle (or average biaxial minerals) ranging from 3 to 3.2.

```r
library(httr)
library(jsonlite)
library (OpenMindat)
```

Initialize OpenMindat

```r
#Please replace the following string “9ce67655d74bcd981e937be80dcea9cb” by your own Mindat token.
mindat_connection("9ce67655d74bcd981e937be80dcea9cb",page_size = 1500)
```

1. Retrieve records of geomaterials with a Mohs hardness greater than 9: 

```r
df1<-geomaterials_hardness_gt(9)
```

```r
df1
```

2. Retrieve records of geomaterials with a Mohs hardness of less than 1: 

```r
df2<-geomaterials_hardness_lt(1)
```

```r
df2
```

3. Retrieve records of geomaterials with a Mohs hardness ranging from 3 to 3.5:

```r
df3<-geomaterials_hardness_range(3,3.5) 
```

```r
df3
```

4. Retrieve records of geomaterials with a density ranging from 3 to 3.2:

```r
df4<-geomaterials_dens_range(3,3.2)
```

```r
df4
```

5. Retrieve records of geomaterials with a calculated 2V angle (or average biaxial minerals) ranging from 3 to 3.2:

```r
df5<-geomaterials_optical2v_range (9,10)
```

```r
df5
```
